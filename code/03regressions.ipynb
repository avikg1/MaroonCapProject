```python
import pandas as pd
from sklearn import linear_model
```

```python

everyday = pd.read_csv('train_data.csv',
        header=0,
        usecols=["Market","Risk_Free", "SMB", "HML", "Long_Rate"])

everyday["SMB_interaction"] = everyday["SMB"] * everyday["Long_Rate"]
everyday["HML_interaction"] = everyday["HML"] * everyday["Long_Rate"]
everyday["M-RF"] = everyday["Market"] - everyday["Risk_Free"]
everyday = everyday.drop('Market', axis=1)
everyday = everyday.drop('Risk_Free', axis=1)
everyday = everyday.drop('Long_Rate', axis=1)
```

```python
dailylist = everyday.values.tolist()
# STOCK SMB HML
interactions = []
returner_df = pd.read_csv("train_data.csv",header=0)
returner_df = returner_df.drop(["Unnamed: 0.1","index","Date","Market", "SMB", "HML", "Long_Rate"], axis=1)

for contender in returner_df.columns:
  if contender == "Risk_Free":
    continue
  returner = returner_df.loc[:, ["Risk_Free", contender]].copy()

  merged_data = pd.merge(returner, everyday, left_index = True, right_index = True, how = "left")

  merged_data = merged_data.dropna(subset = [contender])
  merged_data["Return-Rf"] = merged_data[contender]-merged_data["Risk_Free"]
  merged_data = merged_data.drop(contender, axis=1)
  merged_data = merged_data.drop("Risk_Free", axis=1)
  returnlist = merged_data.values.tolist()
  if contender == "CHRD":
    print(returnlist)
  if len(returnlist) <= 120:
    continue


  y = merged_data[['Return-Rf']].values.tolist()
  x = merged_data.drop('Return-Rf', axis = 1).values.tolist()

  reg = linear_model.LinearRegression()
  reg.fit(x, y)
  interactions.append([contender, reg.coef_[0][2], reg.coef_[0][3]])
result = pd.DataFrame(interactions, columns=["STOCK", "SMB","HML"])
result.to_csv("BIGGERWIN.csv")
```
